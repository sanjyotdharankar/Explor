# **Python Visualization**

## **Matplotlib** 

```python
import matplotlib.pyplot as plt #importing an atplotlib package
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names

```

**About dataset:**
We are studying a built-in dataset called Iris. This dataset is named after a type of flower and focuses on the size of its sepal, which is a leaf-like structure that lies beneath and supports the petals. Sepals protect the developing flower and play a role in attracting pollinators. In this dataset, we can explore the variations in sepal size in iris flowers.

sepals-vs-petals.jpg

```python
#converting in dataframe
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
# Add a new column to the DataFrame for the target variable
df['target'] = iris.target
```

```python
df.head(5)
```

```python
df.plot() #line
```

```python
df.plot(kind='bar') #bar plot
```

```python
df.plot(kind='barh') #bar plot horizontal
```

```python
df.plot(kind='hist') #histogram
```

```python
df.boxplot()
```

```python
# Create the pie chart
grouped = df.groupby('target').size()#only categorical data for pie graph
grouped.plot.pie()
plt.show()
```

```python
#scatter plot of Sepal
df.plot(kind='scatter',x='sepal length (cm)',y='sepal width (cm)')
plt.xlabel('Sepal Length ')
plt.ylabel('Sepal Width ')
plt.title('Scatter plot of Sepal Length vs Sepal Width')
```

```python
#scatter plot for petal
df.plot(kind='scatter',x='petal length (cm)',y='petal width (cm)')
```

```python
plt.imshow(data)
plt.colorbar()
plt.xlabel('Features')
plt.ylabel('Samples')

# set the title of the plot
plt.title('Iris Dataset')

# show the plot
plt.show()
```

```python
#subplot

fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(8,8))
#axs[0,0].scatter(df['petal length (cm)'],df['target'])
axs[0, 0].scatter(df['sepal length (cm)'], df['sepal width (cm)'])
axs[0, 0].set_xlabel('Sepal Length (cm)')
axs[0, 0].set_ylabel('Sepal Width (cm)')

# plot petal length vs petal width
axs[0, 1].scatter(df['sepal length (cm)'], df['petal length (cm)'])
axs[0, 1].set_xlabel('Petal Length (cm)')
axs[0, 1].set_ylabel('Petal Width (cm)')

# plot sepal length vs petal length
axs[1, 0].scatter(df['sepal length (cm)'], df['petal width (cm)'])
axs[1, 0].set_xlabel('Sepal Length (cm)')
axs[1, 0].set_ylabel('Petal Length (cm)')

# plot sepal width vs petal width
axs[1, 1].scatter(df['petal length (cm)'], df['petal width (cm)'])
axs[1, 1].set_xlabel('Sepal Width (cm)')
axs[1, 1].set_ylabel('Petal Width (cm)')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#other way 
# Create a figure and set its size
fig = plt.figure(figsize=(8,8))

# Create a 2x2 grid of subplots
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# Plot the data for each subplot
ax1.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
ax1.set_title('Sepal Length vs Sepal Width')
ax2.scatter(df['sepal length (cm)'], df['petal length (cm)'])
ax2.set_title('Sepal Length vs Petal Length')
ax3.scatter(df['sepal length (cm)'], df['petal width (cm)'])
ax3.set_title('Sepal Length vs Petal Width')
ax4.scatter(df['petal length (cm)'], df['petal width (cm)'])
ax4.set_title('Petal Length vs Petal Width')

# Add a main title 
fig.suptitle('Iris Dataset Scatter Plots', fontsize=14)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

x = [1, 2, 3, 4, 5,9,10]
y = [2, 4, 6, 8, 10,16,20]
plt.plot(x, y)

# Add an annotation to the plot
plt.annotate('This is point (5,10)', xy=(5,10), xytext=(5, 15),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12)
#2. Add a legend to the plot
plt.plot(x, y, label='line 1')

plt.legend(loc='upper left', fontsize=12)
# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 1)
y = x**2
yerr = np.sqrt(x)

plt.errorbar(x, y, yerr=yerr, fmt='-o', color='red', ecolor='black',
             elinewidth=1, capsize=3)

plt.show()
```
